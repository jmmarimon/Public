```python
## Created One-hot encoding for features not binary

x_train_hot=pd.get_dummies(X_train)
x_val_hot= pd.get_dummies(X_val)
```

```python
## Fit the model

random_forest_model = RandomForestClassifier(n_estimators = 1000,
                                             max_depth = 30, 
                                             min_samples_split = 20).fit(x_train_hot,y_train)
```
